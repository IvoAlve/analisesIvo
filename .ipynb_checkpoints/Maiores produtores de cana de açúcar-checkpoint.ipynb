```python
#importando as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#lendo o arquivo
df = pd.read_csv('List of Countries by Sugarcane Production.csv', sep=',', encoding='UTF-8')
df.head()
```

```python
#tamanho do data frame
df.shape
```

```python
#pode-se perceber que teremos transformar alguns valores para numericos
df.info()
```

```python
df.dtypes
```

```python
#renomeando as colunas
df.columns = ["ID", "Pais", "Continente", "Producao(ton)", "Producao por pessoa", "Area cultivada(hec)", "Colheita(kg/hec)"]
```

```python
df.head()
```

```python
df.tail()
```

```python
#tratando valores nulos
df.isnull().sum()
```

```python

sns.heatmap(df.isnull(), cbar = False)
```

```python
df.dropna(inplace = True)
```

```python
#estatisticas
df.describe()
```

```python
df['Continente'].value_counts()
```

```python
#agrupando por continente
agrupado = df.groupby(['Continente']).size()
agrupado
```

```python

plt.figure(figsize=(12,5))
plt.title('Quantidade de pases vendedores de cana de acar por continente')
agrupado.plot.bar(color = 'blue')
plt.xticks(rotation = 45, ha='right');
plt.xlabel('')
#pode-se notar que a frica tem mais pases nessa lista, mas no quer dizer que estes sejam os maiores produtores.
```

```python
df['Producao(ton)'] = df['Producao(ton)'].str.replace('.','', regex = True)
```

```python
df['Area cultivada(hec)'] = df['Area cultivada(hec)'].str.replace('.','', regex = True)
```

```python
df['Producao por pessoa'] = df['Producao por pessoa'].str.replace('.','', regex = True)
```

```python
df['Colheita(kg/hec)'] = df['Colheita(kg/hec)'].str.replace('.','', regex = True)
```

```python
df['Colheita(kg/hec)'] = df['Colheita(kg/hec)'].str.replace(',','', regex = True)
```

```python
df.head()
```

```python
df['Producao(ton)'] = df['Producao(ton)'].astype(int)
```

```python
df['Area cultivada(hec)'] = df['Area cultivada(hec)'].astype(int)
```

```python
df['Colheita(kg/hec)'] = df['Colheita(kg/hec)'].astype(int)
```

```python
df.info()
```

```python
sns.boxplot(df['Producao(ton)'])
#podemos notar vrios outliers visualizando o boxplot
```

```python
maiores = df['Pais'].head(5)
maiores
```

```python
#maiores produtores
plt.pie(df['Producao(ton)'].head(5), labels = maiores[0:5] );
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
sns.barplot(data=df, x='Producao(ton)', y= 'Continente')
plt.title('Maiores produtores por continente')
plt.subplot(1,2,2)
plt.pie(df['Producao(ton)'].head(5), labels = df['Pais'][0:5], autopct='%1.1f%%' )
plt.title('Maiores produtores por pas')
;
```
